```python
knob_weight = 0.1
lr = 0.01

def neural_network(input, weight):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]
input = number_of_toes[0]
true = win_or_lose_binary[0]
prediction = neural_network(input, knob_weight) # output of the network is a single value (scalar)
error = (prediction - true) ** 2
print(error)
```

```python
up_pred = neural_network(input, knob_weight + lr)
up_error = (up_pred - true) ** 2
print(up_error)
```

```python
dn_pred = neural_network(input, knob_weight - lr)
dn_error = (dn_pred - true) ** 2
print(dn_error)
```

```python
if error > up_error or error > dn_error:
    if up_error < dn_error:
        knob_weight += lr
    if up_error > dn_error:
        knob_weight -= lr
pred = neural_network(input, knob_weight)
error = (pred - true) ** 2
print(error)
```

```python
# Neural Network Prediction with hot and cold learning

weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.001

print(f'The weight before iterations is {weight}')
for iteration in range(1102):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2
    print(f'iteration {iteration+1}: Error is {str(error)} and Prediction is {str(prediction)}')

    up_prediction = input * (weight + step_amount)
    up_error = (up_prediction - goal_prediction) ** 2

    dn_prediction = input * (weight - step_amount)
    dn_error = (dn_prediction - goal_prediction) ** 2


    if up_error < dn_error:
        weight += step_amount
    if up_error > dn_error:
        weight -= step_amount
print(f'The weight after iterations is {weight}')
```
