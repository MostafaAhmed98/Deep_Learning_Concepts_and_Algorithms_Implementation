```python
# Training a nueral Network with several inputs and one output
weights = [0.1, 0.2, -0.1]
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]
win_or_lose = [1, 1, 0, 1]
y = win_or_lose[0]
x = [toes[0], wlrec[0], nfans[0]]

def w_sum(a, b):
    assert (len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output

def neural_network(input, weight):
    prediction = w_sum(input, weight)
    return prediction

def delta_sum(number, input):
    output = [0, 0, 0]
    for i in range(len(output)):
        output[i] = number * input[i]
    return output

def applying_alpha(alpha, weights_deltas):
    output = [0, 0, 0]
    for i in range(len(output)):
        output[i] = alpha * weights_deltas[i]
    return output

def updating_weights(weigths, deltas_weights):
    output = [0, 0, 0]
    for i in range(len(output)):
        output[i] = weigths[i] - deltas_weights[i]
    return output

for iteration in range(10):
    pred = neural_network(x, weights)
    error = (pred - y) ** 2
    delta = pred - y
    weights_deltas = delta_sum(delta, x)
    alpha = 0.01
    weights_deltas = applying_alpha(alpha, weights_deltas)
    weights = updating_weights(weights, weights_deltas)
    print(f'Iteration {iteration+1}: ')
    print(f"Error is {error}, Prediction is {pred}")
    print(f'Delta is {delta}')
    print(f"weights is {weights}")
    print(f'Weights deltas is {weights_deltas}')
    print()

```

```python
# Training a nueral Network with one inputs and several output
weights = [0.3, 0.2, 0.9]
wlrec = [0.65, 1.0, 1.0, 0.9]
hurt = [0.1, 0.0, 0.0, 0.1]
win = [1, 1, 0, 1]
sad = [0.1, 0.0, 0.1, 0.2]

x = wlrec[0]
y = [hurt[0], win[0], sad[0]]

def ele_mul(scalar, weights):
    output = [0, 0, 0]
    for i in range(len(weights)):
        output[i] = scalar * weights[i]
    return output

def neural_network(input, weight):
    pred = ele_mul(input, weight)
    return pred

pred = neural_network(x, y)

error = [0,0,0]
deltas = [0,0,0]

for i in range(len(error)):
    error[i] = (pred[i] - y[i]) ** 2
    deltas[i] = pred[i] - y[i]

weights_deltas = [0, 0, 0]

for i in range(len(y)):
    weights_deltas[i] = deltas[i] * x

alpha = 0.1

for i in range(len(weights)):
    weights[i] -= (weights_deltas[i] * alpha)

print(f"Weights: {weights}")
print(f"Weights Delta: {weights_deltas}")
```
